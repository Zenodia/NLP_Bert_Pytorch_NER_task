# clearn the raw tsv dataset to contain no nulls and covert all mis-tagged 0--> O

```python
import pandas as pd
df=pd.read_csv('sv_data/train.tsv', sep='\t')
df = df[df['0'].notnull()]
df.loc[df['0'] == "0", '0'] = "O"
df.to_csv('sv_data/train_no_nulls.tsv', header=False, index=False, sep='\t')
```

```python
def readfile(filename):
    '''
    read file
    '''
    f = open(filename)
    data = []
    sentence = []
    label= []
    for line in f:
        if len(line)==0 or line.startswith('-DOCSTART') or line[0]=="\n":
            if len(sentence) > 0:
                data.append((sentence,label))
                sentence = []
                label = []
            continue
        splits = line.split(' ')
        sentence.append(splits[0])
        label.append(splits[-1][:-1])

    if len(sentence) >0:
        data.append((sentence,label))
        sentence = []
        label = []
    return data
out=readfile('./data/valid.txt')
```

```python
sample_count=0
for s,t in out:
    if sample_count<=2:
        print(len(s),len(t),'\n')
        print(s,t,'\n')
        sample_count+=1
    else:break
```
